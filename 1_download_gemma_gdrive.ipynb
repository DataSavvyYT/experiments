```python
# @title Download Gemma to Google Drive
import os
from google.colab import drive
from huggingface_hub import snapshot_download, login
```

```python
# 1. Mount Google Drive
drive.mount('/content/drive')
```

```python
# 2. Authenticate with Hugging Face
# Enter your token when prompted (make sure you accepted the Gemma license online!)
login()
```

```python
# ---------------------------------------------------------
# CONFIGURATION
# ---------------------------------------------------------
# Use "google/gemma-2-2b-it" for the latest version (Gemma 2)
# Use "google/gemma-2b-it" for the older version (Gemma 1)
MODEL_ID = "google/gemma-2-2b-it"
```

```python
# Path in your Google Drive where the model will be saved
# This creates a folder: MyDrive -> models -> gemma-2-2b-it
DESTINATION_PATH = "/content/drive/MyDrive/models/gemma-2-2b-it"
```

```python
# ---------------------------------------------------------
# DOWNLOAD SCRIPT
# ---------------------------------------------------------
print(f"üöÄ Downloading {MODEL_ID} to {DESTINATION_PATH}...")

try:
    snapshot_download(
        repo_id=MODEL_ID,
        local_dir=DESTINATION_PATH,
        local_dir_use_symlinks=False,  # CRITICAL: Google Drive doesn't support symlinks
        ignore_patterns=["*.h5", "*.ot", "*.msgpack"], # Skip TensorFlow/Flax files to save space
        resume_download=True  # Allows resuming if disconnected
    )
    print(f"‚úÖ Successfully downloaded to: {DESTINATION_PATH}")
    print("You can now load this model in other notebooks using this path!")
except Exception as e:
    print(f"‚ùå Error: {e}")
    print("Common fix: specificy a valid token or accept the license at https://huggingface.co/google/gemma-2-2b-it")
```
